```python
import pandas as pd
import numpy as np
```

```python
Data = pd.read_csv("Data/train.csv", nrows=10_000_000)
```

 # Computing Distance

```python
Data["abs_latitude"]=abs(Data["pickup_latitude"]-Data["dropoff_latitude"])
Data["abs_longitude"]=abs(Data["pickup_longitude"]-Data["dropoff_longitude"])
```

```python
Data["Vector"] = Data["abs_latitude"] + Data["abs_longitude"]
```

```python
Data_na=Data.dropna()
```
